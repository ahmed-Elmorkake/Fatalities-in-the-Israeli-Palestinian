```python
import numpy as np

def fill_ammunition(row):
    if pd.isna(row['ammunition']):
        if row['type_of_injury'] == 'gunfire':
            return 'live ammunition'
        elif row['type_of_injury'] == 'shelling':
            return 'shell'
        elif row['type_of_injury'] == 'explosion':
            return 'bomb'
        elif row['type_of_injury'] == 'stabbing':
            return 'knife'
        elif row['type_of_injury'] == 'hit by a vehicle':
            return 'vehicle'  
        elif row['type_of_injury'] in ['beating', 'physical assault', 'physically assaulted', 'being bludgeoned with an axe', 'Strangulation']:
            return 'Unknown'  
        else:
            return 'Unknown'
    else:
        return row['ammunition']

df['ammunition'] = df.apply(fill_ammunition, axis=1)
```

```python
import pandas as pd


df['age'] = df['age'].fillna(df['age'].median())


categorical_cols = [
    'gender',
    'took_part_in_the_hostilities',
    'place_of_residence',
    'place_of_residence_district',
    'type_of_injury'
]

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")


df['notes'] = df['notes'].fillna("No notes")


print(df.isna().sum())
```
